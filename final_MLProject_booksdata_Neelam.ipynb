```python


# Define a function to remove outliers using z-score
def remove_outliers_zscore(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    return data[(z_scores < threshold).all(axis=1)]

# Define a function to remove outliers using IQR method
def remove_outliers_iqr(data, q1=0.25, q3=0.75):
    q1_values = data.quantile(q1)
    q3_values = data.quantile(q3)
    iqr = q3_values - q1_values
    lower_bound = q1_values - 1.5 * iqr
    upper_bound = q3_values + 1.5 * iqr
    return data[((data > lower_bound) & (data < upper_bound)).all(axis=1)]

# Choose columns with numeric data for outlier removal
numeric_columns = ['bookID', 'authors', 'average_rating', 'language_code', 'num_pages',
       'ratings_count', 'text_reviews_count', 'publisher']

# Remove outliers using z-score method
data_numeric = data[numeric_columns]
data_no_outliers_zscore = remove_outliers_zscore(data_numeric)

# Remove outliers using IQR method
data_no_outliers_iqr = remove_outliers_iqr(data_numeric)

# Print the shape of the original data and the data with outliers removed
print("Original Data Shape:", data.shape)
print("Data Shape (Z-score Outliers Removed):", data_no_outliers_zscore.shape)
print("Data Shape (IQR Outliers Removed):", data_no_outliers_iqr.shape)
```
